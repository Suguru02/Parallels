```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = np.array([1, 2, 4, 7, 8, 16, 20, 40])
gain1 = np.array([0.552745, 1.59985, 3.22799, 4.02715, 5.21158, 6.89138, 12.9174, 14.4466]) 
gain2 = np.array([0.950874, 1.74251, 2.93302, 4.48364, 5.86685, 9.09859, 11.5836, 14.4598]) 
pret_ar = np.array([1.0, 2.0, 4.0, 7.0, 8.0, 16.0, 20.0, 40.0])

# Построение графиков
plt.plot(threads, pret_ar,  'o-', color='black', linewidth=2, markersize=5, label='Ideal Speedup (S = p)')
plt.plot(threads, gain1, 'o-', color='blue', linewidth=2, markersize=5, label='N = 20000')
plt.plot(threads, gain2, 'o-', color='red', linewidth=2, markersize=5, label='N = 40000')

# Подписи и оформление
plt.xlabel('Number of Threads (p)', fontsize=12)
plt.ylabel('Speedup (S)', fontsize=12)
plt.title('Task 1', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)

for i, txt in enumerate(gain1):
    plt.annotate(f'{txt:.1f}', (threads[i], gain1[i]), 
                xytext=(5, -10), textcoords='offset points', fontsize=8, color='blue')
    
for i, txt in enumerate(gain2):
    plt.annotate(f'{txt:.1f}', (threads[i], gain2[i]), 
                xytext=(5, -20), textcoords='offset points', fontsize=8, color='red')

for i, txt in enumerate(pret_ar):
    plt.annotate(f'{txt:.1f}', (threads[i], pret_ar[i]), 
                xytext=(5, 20), textcoords='offset points', fontsize=8, color='black')


plt.show()
```
